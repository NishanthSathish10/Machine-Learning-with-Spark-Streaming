```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv("cluster_data.csv")
```

```python
df
```

```python
batchNo = 1
temp = df[df.batch_no == batchNo]
fig = plt.figure(0, figsize=(7,7))
plt.scatter(temp['pca1'], temp['pca2'], c=temp['y_pred'], cmap='rainbow')
plt.title(f"Batch No : {batchNo}", fontsize=14)
fig = plt.figure(1, figsize=(7,7))
plt.xlabel("pca1")
plt.ylabel("pca2")
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(temp['pca1'],temp['pca2'],temp['pca3'], c=temp['y_pred'], cmap='rainbow')
ax.set_xlabel("pca1")
ax.set_ylabel("pca2")
ax.set_zlabel("pca3")
ax.set_facecolor('white')
plt.title(f"Batch No : {batchNo}", fontsize=14)
```
